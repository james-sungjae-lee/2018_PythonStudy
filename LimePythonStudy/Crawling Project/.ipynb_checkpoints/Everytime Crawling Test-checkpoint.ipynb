## Everytime Crawling Test

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# Using Chrome Webdriver

driver = webdriver.Chrome('/Users/sungjae/Downloads/chromedriver')
driver.implicitly_wait(1)
```

```python
# Login into Everytime

driver.get('https://everytime.kr/login')

driver.find_element_by_name('userid').send_keys('your_everytime_id')
driver.find_element_by_name('password').send_keys('your_everytime_password')

driver.find_element_by_class_name('submit').click()
```

```python
base_url = 'https://everytime.kr/374614/p/'
url = base_url + str(1)
print(url)
```

```python
# go to kookmin university pages

base_url = 'https://everytime.kr/374614/p/'

for i in range(1, 50):
    single_page = base_url + str(i)
    driver.get(single_page)
    
    page_html = driver.page_source
    soup = BeautifulSoup(page_html, 'html.parser')
    notices = soup.select('#container > div.articles > article > a.article > h2')
    
    for n in notices:
        print(n.text.strip())
```
